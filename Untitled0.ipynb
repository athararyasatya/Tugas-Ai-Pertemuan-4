```python
# Langkah 1: Unggah File ke Google Colab
from google.colab import files
uploaded = files.upload()

# Langkah 2: Import Library yang Diperlukan
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Langkah 3: Membaca Dataset
df = pd.read_csv('penjualan_mobil_toyota.csv')

# Langkah 4: Mengubah Kolom 'Bulan' menjadi Tipe Datetime dan Menetapkan sebagai Indeks
df['Bulan'] = pd.to_datetime(df['Bulan'])
df.set_index('Bulan', inplace=True)

# Langkah 5: Visualisasi Tren Penjualan Mobil
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Penjualan'], marker='o')
plt.title('Tren Penjualan Mobil Toyota dari Waktu ke Waktu')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.grid()
plt.show()

# Langkah 6: Instalasi Library Statsmodels
!pip install statsmodels

# Langkah 7: Membangun dan Melatih Model ARIMA
model = ARIMA(df['Penjualan'], order=(1, 1, 1))  # Sesuaikan parameter (p, d, q) sesuai data
model_fit = model.fit()

# Langkah 8: Membuat Prediksi untuk 12 Bulan ke Depan
forecast = model_fit.forecast(steps=12)  # Prediksi 12 bulan ke depan
print("Prediksi Penjualan 12 Bulan ke Depan:")
print(forecast)
```
